```python
ecology = pd.read_csv('../../Body/2Derived/NematodesFromEcologyToGenetics.csv')
```

```python
clc_hosts = pd.read_csv('../../Body/2Derived/CLC_database_hosts.csv')
```

```python
clc_hosts
```

```python
clc_hosts_agg = clc_hosts.groupby(['Parasite.species', 'Parasite.genus', 'Parasite.group']).agg(lambda x: list(x)).reset_index()
```

```python
clc_lifehistory = pd.read_csv('../../Body/2Derived/CLC_database_lifehistory.csv')
```

```python
clc_lifehistory
```

```python
clc_lifehistory_agg = clc_lifehistory.groupby(['Parasite.species', 'Parasite.genus', 'Parasite.group']).agg(lambda x: list(x)).reset_index()
```

```python
nemaplex = pd.read_csv('../../Body/2Derived/nemaplex.csv', sep=';')
```

```python
nemaplex
```

```python
merge_table = ecology.merge(clc_lifehistory_agg, how='left', left_on='Species', right_on='Parasite.species')
```

```python
merge_table = merge_table.merge(clc_hosts_agg, how='left', left_on='Species', right_on='Parasite.species', suffixes=('', '_hosts'))
```

```python
merge_table = merge_table.merge(nemaplex, how='left', left_on='Parasite.genus', right_on='Species', suffixes=('', '_np'))
```

```python
merge_table.head(200)
```

```python
merge_table.to_csv('../../Body/3Results/15_MergingDatabases.csv', sep=';')
```

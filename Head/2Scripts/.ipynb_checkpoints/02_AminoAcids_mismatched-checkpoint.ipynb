```python
import pandas as pd
import os
```

```python
result = pd.DataFrame(columns = ['Unnamed: 0', 'OrganismName', 'GenName', 'GenID', 'TranslatedAminoAcids', 'GenBankAminoAcids'])
result
```

```python
direct = '/home/emulciber/MitoClub/devil-worm/Body/2Derived/aminoacids'
dirname = os.listdir(direct)
```

```python
for file in dirname:
    try:
        df = pd.read_csv((os.path.join(direct, file)), sep='\t')
        result = result.append(df.loc[(df['TranslatedAminoAcids'].apply(lambda x: x[1:]) != df['GenBankAminoAcids'].apply(lambda x: x[1:]))])
    except:
        pass
```

```python
result
```

```python
exclude_list = list(result['OrganismName'])
exclude_list
```

```python
result.to_csv(r'C:\Users\e_mulciber\Desktop\MitoClub\mtDnaInvertebrate\Body\2Derived\AminoAcids_mismatched.csv', sep=';')
```

```python
# ПО КОЛИЧЕСТВУ АМИНОКИСЛОТ

result_q = pd.DataFrame(columns = ['Unnamed: 0', 'OrganismName', 'GenName', 'GenID', 'TranslatedAminoAcids', 'GenBankAminoAcids'])

for file in dirname:
    try:
        df = pd.read_csv((os.path.join(direct, file)), sep='\t')
        result_q = result_q.append(df.loc[(df['TranslatedAminoAcids'].apply(lambda x: len(x)) != df['GenBankAminoAcids'].apply(lambda x: len(x)))])
    except:
        pass
    
result_q
```

```python
result['Q'] = result['TranslatedAminoAcids'].apply(lambda x: len(x)) - result['GenBankAminoAcids'].apply(lambda x: len(x))
```

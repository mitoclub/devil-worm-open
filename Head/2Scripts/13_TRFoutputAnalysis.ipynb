```python
import pandas as pd
from statistics import median
```

```python
trf_output = pd.read_csv('../../Body/2Derived/12_TandemRepeatsOutput.csv')
nematoda = pd.read_csv('../../Body/3Results/10_GenomeStatistics.csv', sep=';')
```

```python
trf_output['Species'] = trf_output['Species'].str.replace('halicephalobus mephisto', 'Halicephalobus mephisto')
trf_output['Species'] = trf_output['Species'].str.replace('Halicephalobus sp.', 'Halicephalobus sp. NKZ332')
```

```python
sum_without_overlaps = []
sum_with_overlaps = []
species_names = []
count_tr = []
t_richness = []

for species in trf_output['Species'].unique():
    
    starts_with = trf_output[trf_output['Species'] == species]['StartsWith'].tolist()
    ends_with = trf_output[trf_output['Species'] == species]['EndsWith'].tolist()
    pattern = trf_output[trf_output['Species'] == species]['Pattern'].tolist()
        

    for i in range(len(starts_with)):
        if i == 0:
            temp_sum = 0
            temp_sum_with = 0
            temp_starts_with = starts_with[i]
            temp_ends_with = ends_with[i]
            t_patterns = []
        
        #         
        temp_sum += ends_with[i] - starts_with[i] + 1
        
        #    
        if starts_with[i] <= temp_ends_with and ends_with[i] >= temp_ends_with:
            temp_ends_with = ends_with[i]
        elif starts_with[i] > temp_ends_with:
            temp_sum_with += temp_ends_with - temp_starts_with + 1
            temp_starts_with = starts_with[i]
            temp_ends_with = ends_with[i]

        if i == (len(starts_with) - 1):
            temp_sum_with += temp_ends_with - temp_starts_with + 1
            
        #  -  ()
        t_patterns.append(pattern[i].count('T') / len(pattern[i]) * 100)
    
            
    species_names.append(species)

    sum_without_overlaps.append(temp_sum)
    sum_with_overlaps.append(temp_sum_with)
    count_tr.append(len(starts_with))
    t_richness.append(median(t_patterns))
```

```python
tr_sums = pd.DataFrame()
tr_sums['Species'] = species_names
tr_sums['TrLengthWithoutOverlaps'] = sum_without_overlaps
tr_sums['TrLengthWithOverlaps'] = sum_with_overlaps
tr_sums['TrCount'] = count_tr
tr_sums['TRichness'] = t_richness
```

```python
nematoda = nematoda.merge(tr_sums, how='left', on='Species')
```

```python
nematoda['TrFractionWithoutOverlaps'] = nematoda['TrLengthWithoutOverlaps'] / nematoda['Length'] * 100
nematoda['TrFractionWithOverlaps'] = nematoda['TrLengthWithOverlaps'] / nematoda['Length'] * 100
```

```python
nematoda.to_csv('../../Body/3Results/13_TRFoutputAnalysis.csv', sep=';', index=False)
```
